```python
def get_country(univer):
    url_base = 'https://en.wikipedia.org/wiki/'

    soup = bs4.BeautifulSoup(requests.get(url_base + univer).content)
    
    try:
        return soup.find('div',{"class": "country-name"}).get_text()
    except:
        return(np.nan)
```

```python
data['Country'] = data['university_name'].apply(lambda x : get_country(x))
```
